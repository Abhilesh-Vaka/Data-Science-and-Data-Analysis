```python
import pandas as pd
d1=pd.read_csv("student-mat.csv",sep=";")
```

```python
#1.Data Loading
d1.head()
```

```python
#2.Data Exploration
#To Check the No of null values in the Dataset
d1.isnull().sum()
```

```python
#To check the Data type of Every Column
d1.dtypes
```

```python
#To check the size of Dataset(row,col)
d1.shape
```

```python
#3.Data Cleaning
d1.dropna()
```

```python
#As There are no Missing Values data cleaning Doesn't change any values of the Data Set
d1['Medu']=d1['Medu'].fillna(d1['Medu'].median())
d1.head()
```

```python
#Dropping Duplicate values
d1.drop_duplicates()
```

```python
#4.Data Analysis Questions
#Qn-01
avg_score=round(d1['G3'].mean(),2)
print(f'The Average Score of Students in Math is :{avg_score}')
```

```python
#Qn-02
student_above_15=(d1['G3']>15).sum()
print(f'Students Scored above 15 in their Final Grade are :{student_above_15}')
```

```python
#Qn-03
correlation=round(d1['studytime'].corr(d1['G3']),6)
print(f'The Correlation between study time of a student annd their final grades is:{correlation}')
```

```python
gender=d1.groupby('sex')['G3'].mean()
highest_averaged_gender = gender.idxmax()
print(f'Averages of each gender is:{gender}')
print(f"The Gender with the highest average final grade (G3): {highest_averaged_gender}")
```

```python
#5.Data Visualisation
# Plot a histogram of final grades (G3)
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(d1['G3'], bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(d1['studytime'], d1['G3'], alpha=0.6, color='green')
plt.title('Scatter Plot: Study Time vs Final Grade (G3)')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
#Bar chart comparing the average scores of male and female students.
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.barplot(x=gender.index, y=gender.values, palette='pastel', hue=gender.index)
plt.title('Average Final Grade (G3) by Gender (M/F)')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```
