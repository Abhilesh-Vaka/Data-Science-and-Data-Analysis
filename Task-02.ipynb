```python
#Project-01
#1.Dataset Selection
#Choosing "Global Superstore" dataset containing columns such as Sales, Profit, Region, and Product Categories
import pandas as pd
```

```python
#Reading the excel Dataset
dt=pd.read_excel('global_superstore_2016.xlsx')
dt.head()
```

```python
#Checking for the null values
dt.isnull().sum()
dt.shape
```

```python
#Cleaning Data:Handle missing values by filling them with appropriate measures (mean, median, or placeholders) or by removing affected rows/columns.
#Here only postal codes have missing values,but we cant replace postal codes with mean,
#median or mode as it may represent wrong address.so replacing null values using '00000'
dt['Postal Code']=dt['Postal Code'].fillna('00000')
```

```python
print(dt['Postal Code'])
```

```python
#Remove duplicates
dt.drop_duplicates()
```

```python
#Detect and handle outliers using IQR(interquartile range) statistical technique
sales_data=dt['Sales']
Q1=sales_data.quantile(0.25)
Q3=sales_data.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

#Identifying Outliers
outliers=dt[(sales_data<lower_bound)|(sales_data>upper_bound)]
print('Outliers Detected/n',outliers)

#Data with No Outliers
dt_no_outliers=dt[(sales_data>=lower_bound)&(sales_data<=upper_bound)]
print(f'Dataset Without Outliers:{dt_no_outliers}')

#replacing outliers with median values
median_sales = sales_data.median()
dt['Sales'] = dt['Sales'].apply(lambda x: median_sales if (x < lower_bound) or (x > upper_bound) else x)

# Display the updated data with replaced outliers
print("\nData with Outliers Replaced by Median:\n", dt[['Order ID', 'Sales', 'Profit']].head())
```

```python
#Statistical Analysis
#applying  mean, median, standard deviation, and variance to understand Distribution across Sales,Profit & Shipping Cost
#Mean 
profit_data=dt['Profit']
shipping_cost=dt['Shipping Cost']
mean_sales=sales_data.mean()
mean_profit=profit_data.mean()
mean_shipping_cost=shipping_cost.mean()
print('MEAN OF THE DATA:')
print('Mean of Sales ($):',round(mean_sales,2))
print('Mean of profits ($):',round(mean_profit,2))
print('Mean of Shipping Cost ($):',round(mean_shipping_cost,2),'\n')

#Median
median_sales=sales_data.median()
median_profit=profit_data.median()
median_shipping_cost=shipping_cost.median()
print('MEDIAN OF THE DATA:')
print('Median of Sales ($):',round(median_sales,2))
print('Median of profits ($):',round(median_profit,2))
print('Median of Shipping Cost ($):',round(median_shipping_cost,2),'\n')
print('"We can observe that there is a huge difference between Mean and Median of the Data (Mean>Median),This Indicates that there are more Number of Low to Moderate Sales But only few Large Corporate/Bulk Orders has been placed.That is the Reason there is a Difference Between Mean and Median"','\n')
#Mode
mode_sales=sales_data.mode()
mode_profit=profit_data.mode()
mode_shipping_cost=shipping_cost.mode()
print('MODE OF THE DATA:')
print('Mode of Sales :',mode_sales)
print('Mode of profits :',mode_profit)
print('Mode of Shipping Cost :',mode_shipping_cost,'\n')

#Standard Deviation
std_sales=sales_data.std()
std_profit=profit_data.std()
std_shipping_cost=shipping_cost.std()
print('STANDARD DEVIATION OF THE DATA:')
print('Standard Deviation of Sales :',round(std_sales,2))
print('Standard Deviation of profits :',round(std_profit,2))
print('Standard Deviation of Shipping Cost :',round(mode_shipping_cost,2),'\n')

#Variance
var_sales=sales_data.var()
var_profit=profit_data.var()
var_shipping_cost=shipping_cost.var()
print('VARIANCE OF THE DATA:')
print('Variance of Sales :',round(var_sales,2))
print('Variance of profits :',round(var_profit,2))
print('Variance of Shipping Cost :',round(var_shipping_cost,2))
```

```python
discount=dt['Discount']
df=pd.DataFrame({
    'Sales':sales_data,
    'Profits':profit_data,
    'Discount':discount,
    'Shipping Cost':shipping_cost
})
correlation_matrix=df.corr()
print('Correlation Between the Data:\n',correlation_matrix)
print('\nThe Relation Btwn Sales and profits is positively correlated.Whenever Sales Increase Profits tend to increse Slowly')
print('\nThe Relation Btwn Sales and discounts is negatively correlated.The reason to increase sales is dependent upon the reduction of discounts over the Products')
```
